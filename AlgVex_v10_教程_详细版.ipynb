### æ–¹æ³•2: æ‰‹åŠ¨è·å–

ç”±äºç½‘ç»œé™åˆ¶ï¼Œä¸‹é¢ä½¿ç”¨æ¨¡æ‹Ÿæ•°æ®æ¼”ç¤ºå®Œæ•´æµç¨‹ï¼š

````python
# ============================================================
#            ğŸ“¥ ç”Ÿæˆæ¨¡æ‹Ÿæ•°æ® (å®Œæ•´ç‰ˆ)
# ============================================================
# æ™ºèƒ½æ£€æµ‹: å¦‚æœæ•°æ®å·²å­˜åœ¨åˆ™è·³è¿‡ç”Ÿæˆ
# ============================================================

import numpy as np
import pandas as pd
from pathlib import Path

# é…ç½®
DATA_DIR = Path.home() / ".algvex" / "data" / "1h"
DATA_DIR.mkdir(parents=True, exist_ok=True)

SYMBOLS = ["btcusdt", "ethusdt"]

print("ğŸ” æ£€æŸ¥æ•°æ®æ–‡ä»¶çŠ¶æ€...")
print("=" * 60)

# æ£€æŸ¥å“ªäº›æ•°æ®éœ€è¦ç”Ÿæˆ
existing = []
to_generate = []

for symbol in SYMBOLS:
    file_path = DATA_DIR / f"{symbol}.parquet"
    if file_path.exists():
        df = pd.read_parquet(file_path)
        existing.append((symbol, len(df)))
        print(f"  âœ… {symbol}.parquet - å·²å­˜åœ¨ ({len(df):,} æ¡è®°å½•)")
    else:
        to_generate.append(symbol)
        print(f"  â³ {symbol}.parquet - å¾…ç”Ÿæˆ")

print()

if not to_generate:
    print("=" * 60)
    print("ğŸ‰ æ‰€æœ‰æ•°æ®æ–‡ä»¶å·²å­˜åœ¨ï¼Œæ— éœ€é‡æ–°ç”Ÿæˆ!")
    print("=" * 60)
    print("\nğŸ’¡ å¦‚éœ€é‡æ–°ç”Ÿæˆï¼Œè¯·å…ˆåˆ é™¤ç°æœ‰æ–‡ä»¶:")
    print(f"   {DATA_DIR}")
else:
    print(f"ğŸ“¥ éœ€è¦ç”Ÿæˆ {len(to_generate)} ä¸ªæ•°æ®æ–‡ä»¶...")
    print("=" * 60)
    
    # ç”Ÿæˆ 2 å¹´çš„ 1h æ•°æ®
    np.random.seed(42)
    n_hours = 24 * 365 * 2
    
    def generate_ohlcv(base_price, n, volatility=0.02):
        """ç”Ÿæˆ OHLCV æ•°æ®"""
        returns = np.random.normal(0.0001, volatility, n)
        trend = np.sin(np.linspace(0, 8*np.pi, n)) * 0.001
        returns = returns + trend
        close = base_price * np.exp(np.cumsum(returns))
        
        intrabar_vol = volatility * 0.5
        high = close * (1 + np.abs(np.random.normal(0, intrabar_vol, n)))
        low = close * (1 - np.abs(np.random.normal(0, intrabar_vol, n)))
        open_price = np.roll(close, 1)
        open_price[0] = base_price
        
        volume = np.random.lognormal(6, 0.5, n)
        
        dates = pd.date_range("2023-01-01", periods=n, freq="1h", tz="UTC")
        
        return pd.DataFrame({
            "open": open_price,
            "high": high,
            "low": low,
            "close": close,
            "volume": volume
        }, index=dates)
    
    configs = {"btcusdt": 42000, "ethusdt": 2800}
    
    for symbol in to_generate:
        if symbol in configs:
            print(f"\n   ç”Ÿæˆ {symbol} æ•°æ®...")
            df = generate_ohlcv(configs[symbol], n_hours)
            df.to_parquet(DATA_DIR / f"{symbol}.parquet")
            print(f"   âœ… {symbol}.parquet ({len(df):,} æ¡)")
    
    print()
    print("=" * 60)
    print("ğŸ‰ æ•°æ®ç”Ÿæˆå®Œæˆ!")
    print("=" * 60)

# æ˜¾ç¤ºæ•°æ®ç›®å½•å†…å®¹
print(f"\nğŸ“‚ æ•°æ®ç›®å½•: {DATA_DIR}")
for f in DATA_DIR.glob("*.parquet"):
    size = f.stat().st_size / 1024 / 1024
    print(f"   ğŸ“„ {f.name} ({size:.2f} MB)")
````

````python
# ============================================================
#            ğŸ“ˆ æ•°æ®å¯è§†åŒ–è¯´æ˜
# ============================================================
# æœ¬æ•™ç¨‹ä½¿ç”¨ Plotly è¿›è¡Œäº¤äº’å¼å¯è§†åŒ–
# å¯è§†åŒ–ä»…ç”¨äºå±•ç¤ºæ•°æ®ï¼Œä¸æ¶‰åŠç³»ç»Ÿæ ¸å¿ƒåŠŸèƒ½
#
# å®‰è£…: pip install plotly
# ============================================================

import plotly.graph_objs as go
from plotly.subplots import make_subplots
import plotly.io as pio

pio.renderers.default = "notebook"

print("ğŸ“Š å¯è§†åŒ–åº“å·²åŠ è½½")
print("   - Plotly: äº¤äº’å¼å›¾è¡¨")
print("   - æ”¯æŒç¼©æ”¾ã€å¹³ç§»ã€æ‚¬åœæŸ¥çœ‹æ•°æ®ç‚¹")
````

````python
# ============================================================
#            ğŸ” æ•°æ®è´¨é‡æ£€æŸ¥ (è¯¦ç»†ç‰ˆ)
# ============================================================

import pandas as pd
from pathlib import Path

DATA_DIR = Path.home() / ".algvex" / "data" / "1h"

def check_data_quality(file_path):
    """å…¨é¢çš„æ•°æ®è´¨é‡æ£€æŸ¥"""
    
    df = pd.read_parquet(file_path)
    name = file_path.stem
    
    print(f"\n{'='*60}")
    print(f"ğŸ“Š {name.upper()} æ•°æ®è´¨é‡æŠ¥å‘Š")
    print(f"{'='*60}")
    
    issues = []
    
    # 1. åŸºæœ¬ä¿¡æ¯
    print(f"\nğŸ“ˆ åŸºæœ¬ä¿¡æ¯:")
    print(f"   è¡Œæ•°: {len(df):,}")
    print(f"   æ—¶é—´èŒƒå›´: {df.index.min()} ~ {df.index.max()}")
    print(f"   æ—¶åŒº: {df.index.tz}")
    
    # 2. ç¼ºå¤±å€¼æ£€æŸ¥
    print(f"\nğŸ” ç¼ºå¤±å€¼æ£€æŸ¥:")
    null_counts = df.isnull().sum()
    if null_counts.sum() == 0:
        print("   âœ… æ— ç¼ºå¤±å€¼")
    else:
        for col, count in null_counts.items():
            if count > 0:
                print(f"   âŒ {col}: {count} ä¸ªç¼ºå¤±å€¼")
                issues.append(f"{col} æœ‰ {count} ä¸ªç¼ºå¤±å€¼")
    
    # 3. é‡å¤è¡Œæ£€æŸ¥
    print(f"\nğŸ” é‡å¤æ£€æŸ¥:")
    dup_index = df.index.duplicated().sum()
    if dup_index == 0:
        print("   âœ… æ— é‡å¤æ—¶é—´æˆ³")
    else:
        print(f"   âŒ {dup_index} ä¸ªé‡å¤æ—¶é—´æˆ³")
        issues.append(f"{dup_index} ä¸ªé‡å¤æ—¶é—´æˆ³")
    
    # 4. æ—¶é—´é¡ºåºæ£€æŸ¥
    print(f"\nğŸ” æ—¶é—´é¡ºåº:")
    if df.index.is_monotonic_increasing:
        print("   âœ… æ—¶é—´å‡åºæ’åˆ—")
    else:
        print("   âŒ æ—¶é—´é¡ºåºé”™ä¹±")
        issues.append("æ—¶é—´é¡ºåºé”™ä¹±")
    
    # 5. OHLC å…³ç³»æ£€æŸ¥
    print(f"\nğŸ” OHLC å…³ç³»:")
    ohlc_valid = (
        (df['high'] >= df['open']) & 
        (df['high'] >= df['close']) &
        (df['low'] <= df['open']) & 
        (df['low'] <= df['close']) &
        (df['high'] >= df['low'])
    )
    invalid_count = (~ohlc_valid).sum()
    if invalid_count == 0:
        print("   âœ… OHLC å…³ç³»æ­£ç¡®")
    else:
        print(f"   âŒ {invalid_count} è¡Œ OHLC å…³ç³»å¼‚å¸¸")
        issues.append(f"{invalid_count} è¡Œ OHLC å…³ç³»å¼‚å¸¸")
    
    # 6. ä»·æ ¼åˆç†æ€§æ£€æŸ¥
    print(f"\nğŸ” ä»·æ ¼åˆç†æ€§:")
    negative_prices = (df[['open', 'high', 'low', 'close']] <= 0).any(axis=1).sum()
    if negative_prices == 0:
        print("   âœ… ä»·æ ¼å…¨éƒ¨ä¸ºæ­£")
    else:
        print(f"   âŒ {negative_prices} è¡Œæœ‰éæ­£ä»·æ ¼")
        issues.append(f"{negative_prices} è¡Œæœ‰éæ­£ä»·æ ¼")
    
    # 7. ç»Ÿè®¡æ‘˜è¦
    print(f"\nğŸ“Š ç»Ÿè®¡æ‘˜è¦:")
    print(f"   å¼€ç›˜ä»·: ${df['open'].mean():,.2f} (å‡å€¼)")
    print(f"   æœ€é«˜ä»·: ${df['high'].max():,.2f} (æœ€å¤§)")
    print(f"   æœ€ä½ä»·: ${df['low'].min():,.2f} (æœ€å°)")
    print(f"   æˆäº¤é‡: {df['volume'].mean():,.2f} (æ—¥å‡)")
    
    # 8. ç»“è®º
    print(f"\n{'='*60}")
    if len(issues) == 0:
        print("âœ… æ•°æ®è´¨é‡æ£€æŸ¥é€šè¿‡!")
    else:
        print(f"âš ï¸ å‘ç° {len(issues)} ä¸ªé—®é¢˜:")
        for issue in issues:
            print(f"   - {issue}")
    
    return len(issues) == 0

# æ£€æŸ¥æ‰€æœ‰æ•°æ®æ–‡ä»¶
all_passed = True
for file_path in DATA_DIR.glob("*.parquet"):
    passed = check_data_quality(file_path)
    all_passed = all_passed and passed

print("\n" + "=" * 60)
if all_passed:
    print("ğŸ‰ æ‰€æœ‰æ•°æ®æ–‡ä»¶è´¨é‡æ£€æŸ¥é€šè¿‡!")
else:
    print("âš ï¸ éƒ¨åˆ†æ•°æ®æ–‡ä»¶å­˜åœ¨é—®é¢˜ï¼Œè¯·æ£€æŸ¥ä¿®å¤")
````

## 3.2 KBAR ç±»å› å­ (Kçº¿å½¢æ€ï¼Œ9ä¸ª)

KBAR å› å­æè¿°å•æ ¹Kçº¿çš„å½¢æ€ç‰¹å¾ï¼š

### å…¬å¼å®šä¹‰

```python
# åŸºç¡€å˜é‡
O = open   # å¼€ç›˜ä»·
H = high   # æœ€é«˜ä»·
L = low    # æœ€ä½ä»·
C = close  # æ”¶ç›˜ä»·

# KBAR å› å­
KMID  = (C - O) / O                      # Kçº¿å®ä½“ä¸­å¿ƒåç§»
KLEN  = (H - L) / O                      # Kçº¿æ€»é•¿åº¦
KMID2 = (C - O) / (H - L + Îµ)            # å®ä½“ç›¸å¯¹äºå½±çº¿çš„æ¯”ä¾‹
KUP   = (H - max(O, C)) / O              # ä¸Šå½±çº¿é•¿åº¦
KUP2  = (H - max(O, C)) / (H - L + Îµ)    # ä¸Šå½±çº¿ç›¸å¯¹æ¯”ä¾‹
KLOW  = (min(O, C) - L) / O              # ä¸‹å½±çº¿é•¿åº¦  
KLOW2 = (min(O, C) - L) / (H - L + Îµ)    # ä¸‹å½±çº¿ç›¸å¯¹æ¯”ä¾‹
KSFT  = (2C - H - L) / O                 # æ”¶ç›˜åç§»
KSFT2 = (2C - H - L) / (H - L + Îµ)       # æ”¶ç›˜åç§»æ¯”ä¾‹
```

### å«ä¹‰è§£è¯»

| å› å­ | å«ä¹‰ | å¤šå¤´ä¿¡å· | ç©ºå¤´ä¿¡å· |
|------|------|----------|----------|
| KMID | æ¶¨è·Œå¹… | æ­£å€¼ (ä¸Šæ¶¨) | è´Ÿå€¼ (ä¸‹è·Œ) |
| KLEN | æŒ¯å¹… | å¤§å€¼ (æ´»è·ƒ) | å°å€¼ (ç›˜æ•´) |
| KUP | ä¸Šå½±çº¿ | å°å€¼ (ä¹°å‹å¼º) | å¤§å€¼ (å–å‹å¼º) |
| KLOW | ä¸‹å½±çº¿ | å¤§å€¼ (ä¹°ç›˜å¼º) | å°å€¼ (å–ç›˜å¼º) |
| KSFT | æ”¶ç›˜ä½ç½® | æ­£å€¼ (åä¸Š) | è´Ÿå€¼ (åä¸‹) |

````python
# ============================================================
#            ğŸ“Š KBAR å› å­è®¡ç®—æ¼”ç¤º
# ============================================================
import numpy as np
import pandas as pd
from pathlib import Path

DATA_DIR = Path.home() / ".algvex" / "data" / "1h"

# è¯»å– BTC æ•°æ®
btc_df = pd.read_parquet(DATA_DIR / "btcusdt.parquet")

# å–æœ€è¿‘ 100 æ¡æ•°æ®æ¼”ç¤º
df = btc_df.tail(100).copy()

# è®¡ç®— KBAR å› å­
close = df["close"]
open_ = df["open"]
high = df["high"]
low = df["low"]

eps = 1e-12  # é¿å…é™¤é›¶

kbar = pd.DataFrame(index=df.index)
kbar["KMID"] = (close - open_) / open_
kbar["KLEN"] = (high - low) / open_
kbar["KMID2"] = (close - open_) / (high - low + eps)
kbar["KUP"] = (high - np.maximum(open_, close)) / open_
kbar["KUP2"] = (high - np.maximum(open_, close)) / (high - low + eps)
kbar["KLOW"] = (np.minimum(open_, close) - low) / open_
kbar["KLOW2"] = (np.minimum(open_, close) - low) / (high - low + eps)
kbar["KSFT"] = (2 * close - high - low) / open_
kbar["KSFT2"] = (2 * close - high - low) / (high - low + eps)

print("ğŸ“Š KBAR å› å­ç¤ºä¾‹ (æœ€è¿‘ 5 æ¡)")
print("=" * 80)
print(kbar.tail().round(4).to_string())

print("\nğŸ“ˆ KBAR å› å­ç»Ÿè®¡")
print("=" * 80)
print(kbar.describe().round(4).to_string())
````

### ROC å› å­ (åŠ¨é‡)
```python
ROC{d} = close / close.shift(d) - 1
```
- **å«ä¹‰**: è¿‡å» d æ ¹Kçº¿çš„æ¶¨è·Œå¹…
- **å¤šå¤´ä¿¡å·**: ROC > 0 (ä¸Šæ¶¨è¶‹åŠ¿)
- **ç©ºå¤´ä¿¡å·**: ROC < 0 (ä¸‹è·Œè¶‹åŠ¿)

### MA å› å­ (å‡çº¿åç¦»)
```python
MA{d} = close / close.rolling(d).mean() - 1
```
- **å«ä¹‰**: å½“å‰ä»·æ ¼ç›¸å¯¹äº d å‘¨æœŸå‡çº¿çš„åç¦»
- **å¤šå¤´ä¿¡å·**: MA > 0 (ä»·æ ¼åœ¨å‡çº¿ä¸Šæ–¹)
- **ç©ºå¤´ä¿¡å·**: MA < 0 (ä»·æ ¼åœ¨å‡çº¿ä¸‹æ–¹)

### STD å› å­ (æ³¢åŠ¨ç‡)
```python
STD{d} = close.rolling(d).std() / close
```
- **å«ä¹‰**: è¿‡å» d å‘¨æœŸçš„æ ‡å‡†åŒ–æ³¢åŠ¨ç‡
- **é«˜å€¼**: å¸‚åœºæ´»è·ƒï¼Œå¯èƒ½æœ‰è¶‹åŠ¿
- **ä½å€¼**: å¸‚åœºå¹³é™ï¼Œç›˜æ•´çŠ¶æ€

### MAX/MIN å› å­ (çªç ´)
```python
MAX{d} = close / high.rolling(d).max() - 1
MIN{d} = close / low.rolling(d).min() - 1
```
- **MAX æ¥è¿‘ 0**: æ¥è¿‘è¿‘æœŸé«˜ç‚¹ (å¯èƒ½çªç ´)
- **MIN æ¥è¿‘ 0**: æ¥è¿‘è¿‘æœŸä½ç‚¹ (å¯èƒ½ç ´ä½)

### QTLU/QTLD å› å­ (åˆ†ä½æ•°)
```python
QTLU{d} = close / close.rolling(d).quantile(0.8) - 1
QTLD{d} = close / close.rolling(d).quantile(0.2) - 1
```
- **QTLU > 0**: ä»·æ ¼é«˜äº 80% åˆ†ä½ (å¼ºåŠ¿)
- **QTLD < 0**: ä»·æ ¼ä½äº 20% åˆ†ä½ (å¼±åŠ¿)

### RSV å› å­ (ç›¸å¯¹å¼ºå¼±ä½ç½®)
```python
RSV{d} = (close - low.rolling(d).min()) / (high.rolling(d).max() - low.rolling(d).min() + Îµ)
```
- **RSV > 0.8**: è¶…ä¹°åŒºåŸŸ
- **RSV < 0.2**: è¶…å–åŒºåŸŸ

### CORR/CORD å› å­ (ä»·é‡ç›¸å…³)
```python
CORR{d} = close.rolling(d).corr(volume)
CORD{d} = close.pct_change().rolling(d).corr(volume.pct_change())
```
- **æ­£ç›¸å…³**: é‡ä»·é½å‡/é½è·Œ (è¶‹åŠ¿ç¡®è®¤)
- **è´Ÿç›¸å…³**: é‡ä»·èƒŒç¦» (å¯èƒ½åè½¬)

### ä¸ºä»€ä¹ˆéœ€è¦å½’ä¸€åŒ–?

3. **å¼‚å¸¸å€¼å¤„ç†**: Z-score å¯ä»¥å‡¸æ˜¾å¼‚å¸¸å€¼

### å½’ä¸€åŒ–å…¬å¼

```python

```

- `mean`: è®­ç»ƒé›†çš„åˆ—å‡å€¼
- `std`: è®­ç»ƒé›†çš„åˆ—æ ‡å‡†å·®

### é˜²æ³„æ¼æœºåˆ¶

```
âš ï¸ é‡è¦: éªŒè¯é›†/æµ‹è¯•é›†/å®ç›˜ å¿…é¡»ä½¿ç”¨è®­ç»ƒé›†çš„ç»Ÿè®¡é‡!
```

### strict æ¨¡å¼

| æ¨¡å¼ | ç¼ºå¤±åˆ— | NaN å€¼ | ç”¨é€” |
|------|--------|--------|------|
| `strict=True` | æŠ›å¼‚å¸¸ | æŠ›å¼‚å¸¸ | å›æµ‹/å®ç›˜ |

````python
# ============================================================
#            ğŸ”§ å½’ä¸€åŒ–å™¨æ¼”ç¤º
# ============================================================
import sys
import os
from scripts.unified_features import FeatureNormalizer, compute_unified_features
from pathlib import Path
import pandas as pd
import numpy as np

DATA_DIR = Path.home() / ".algvex" / "data" / "1h"

# è¯»å–æ•°æ®
btc_df = pd.read_parquet(DATA_DIR / "btcusdt.parquet")

# è®¡ç®—ç‰¹å¾
features = compute_unified_features(btc_df)
features_valid = features.dropna()

# æ¨¡æ‹Ÿè®­ç»ƒ/éªŒè¯é›†åˆ’åˆ†
n = len(features_valid)
train_size = int(n * 0.7)

X_train = features_valid.iloc[:train_size]
X_valid = features_valid.iloc[train_size:]

print("ğŸ“Š å½’ä¸€åŒ–å™¨æ¼”ç¤º")
print("=" * 60)
print(f"è®­ç»ƒé›†: {len(X_train)} æ¡")
print(f"éªŒè¯é›†: {len(X_valid)} æ¡")

# åˆ›å»ºå½’ä¸€åŒ–å™¨
normalizer = FeatureNormalizer()

# åœ¨è®­ç»ƒé›†ä¸Š fit_transform
X_train_norm = normalizer.fit_transform(X_train)

print(f"\nâœ… å½’ä¸€åŒ–å™¨å·²åœ¨è®­ç»ƒé›†ä¸Šæ‹Ÿåˆ")
print(f"   ä¿å­˜çš„ç»Ÿè®¡é‡: {len(normalizer.mean)} åˆ—")

# å½’ä¸€åŒ–å‰åå¯¹æ¯”
print("\nğŸ“ˆ å½’ä¸€åŒ–å‰åå¯¹æ¯” (å‰ 5 åˆ—):")
print("-" * 60)
cols = X_train.columns[:5]
print("å½’ä¸€åŒ–å‰ (mean):")
print(X_train[cols].mean().round(6).to_string())
print("\nå½’ä¸€åŒ–å (mean â‰ˆ 0):")
print(X_train_norm[cols].mean().round(6).to_string())

# åœ¨éªŒè¯é›†ä¸Š transform (strict=True)
print("\nğŸ”’ åœ¨éªŒè¯é›†ä¸Šä½¿ç”¨ strict=True æ¨¡å¼:")
try:
    X_valid_norm = normalizer.transform(X_valid, strict=True)
    print(f"   âœ… æˆåŠŸ! å½’ä¸€åŒ–å shape: {X_valid_norm.shape}")
except ValueError as e:
    print(f"   âŒ å¤±è´¥: {e}")
````

````python
# ============================================================
#            ğŸ¯ å®Œæ•´æ¨¡å‹è®­ç»ƒæµç¨‹
# ============================================================
import sys
import os
import json
import pickle
from pathlib import Path
from datetime import datetime

import numpy as np
import pandas as pd
import lightgbm as lgb

from scripts.unified_features import (
    compute_unified_features, 
    compute_label,
    FeatureNormalizer, 
    FEATURE_COLUMNS
)

print("ğŸ¯ AlgVex æ¨¡å‹è®­ç»ƒ")
print("=" * 60)

# ========== é…ç½® ==========
DATA_DIR = Path.home() / ".algvex" / "data" / "1h"
MODEL_DIR = Path.home() / ".algvex" / "models" / "qlib_alpha"
MODEL_DIR.mkdir(parents=True, exist_ok=True)

INSTRUMENTS = ["btcusdt", "ethusdt"]

# æ—¶é—´é…ç½® (æ³¨æ„ gap)
TRAIN_START = "2023-01-01"
TRAIN_END = "2024-06-15"    # è®­ç»ƒç»“æŸ
VALID_START = "2024-07-01"  # éªŒè¯å¼€å§‹ (ç•™ 2 å‘¨ gap)
VALID_END = "2024-12-31"

print(f"ğŸ“ æ•°æ®ç›®å½•: {DATA_DIR}")
print(f"ğŸ“ æ¨¡å‹ç›®å½•: {MODEL_DIR}")
print(f"ğŸ“Š äº¤æ˜“å¯¹: {INSTRUMENTS}")
print(f"ğŸ“… è®­ç»ƒæœŸ: {TRAIN_START} ~ {TRAIN_END}")
print(f"ğŸ“… éªŒè¯æœŸ: {VALID_START} ~ {VALID_END}")

# ========== åŠ è½½æ•°æ® ==========
print("\n" + "=" * 60)
print("ğŸ“¥ åŠ è½½æ•°æ®...")

all_features = []
all_labels = []

for inst in INSTRUMENTS:
    file_path = DATA_DIR / f"{inst}.parquet"
    if not file_path.exists():
        print(f"   âš ï¸ è·³è¿‡ {inst}: æ–‡ä»¶ä¸å­˜åœ¨")
        continue
    
    df = pd.read_parquet(file_path)
    print(f"   âœ… {inst}: {len(df):,} æ¡")
    
    # è®¡ç®—ç‰¹å¾å’Œæ ‡ç­¾
    features = compute_unified_features(df)
    labels = compute_label(df)
    
    # æ·»åŠ å…ƒä¿¡æ¯
    features["datetime"] = features.index
    features["instrument"] = inst
    labels.name = "label"
    
    all_features.append(features)
    all_labels.append(labels)

# åˆå¹¶æ•°æ®
features_df = pd.concat(all_features)
labels_df = pd.concat(all_labels)

data = features_df.copy()
data["label"] = labels_df.values
data = data.dropna()

print(f"\nâœ… æ€»æ•°æ®é‡: {len(data):,} æ¡")
````

````python
# ============================================================
#            ğŸ“Š æ•°æ®åˆ‡åˆ†ä¸å½’ä¸€åŒ–
# ============================================================

# è§£ææ—¶é—´
train_start_ts = pd.Timestamp(TRAIN_START, tz="UTC")
train_end_ts = pd.Timestamp(TRAIN_END, tz="UTC") + pd.Timedelta(hours=23, minutes=59, seconds=59)
valid_start_ts = pd.Timestamp(VALID_START, tz="UTC")
valid_end_ts = pd.Timestamp(VALID_END, tz="UTC") + pd.Timedelta(hours=23, minutes=59, seconds=59)

# åˆ‡åˆ†æ•°æ®é›†
train_mask = (data["datetime"] >= train_start_ts) & (data["datetime"] <= train_end_ts)
valid_mask = (data["datetime"] >= valid_start_ts) & (data["datetime"] <= valid_end_ts)

X_train = data.loc[train_mask, FEATURE_COLUMNS]
y_train = data.loc[train_mask, "label"]
X_valid = data.loc[valid_mask, FEATURE_COLUMNS]
y_valid = data.loc[valid_mask, "label"]

print("ğŸ“Š æ•°æ®é›†åˆ’åˆ†")
print("=" * 60)
print(f"è®­ç»ƒé›†: {len(X_train):,} æ¡ ({TRAIN_START} ~ {TRAIN_END})")
print(f"éªŒè¯é›†: {len(X_valid):,} æ¡ ({VALID_START} ~ {VALID_END})")

# æ ‡ç­¾åˆ†å¸ƒ
print("\nğŸ“ˆ æ ‡ç­¾åˆ†å¸ƒ (è®­ç»ƒé›†):")
print(f"   æ­£æ”¶ç›Š (>0): {(y_train > 0).sum():,} ({(y_train > 0).mean():.1%})")
print(f"   è´Ÿæ”¶ç›Š (<0): {(y_train < 0).sum():,} ({(y_train < 0).mean():.1%})")
print(f"   å‡å€¼: {y_train.mean():.6f}")
print(f"   æ ‡å‡†å·®: {y_train.std():.6f}")

# å½’ä¸€åŒ– (åªç”¨è®­ç»ƒé›† fit)
print("\nğŸ”§ å½’ä¸€åŒ–ç‰¹å¾...")
normalizer = FeatureNormalizer()
X_train_norm = normalizer.fit_transform(X_train)
X_valid_norm = normalizer.transform(X_valid, strict=False)

print("   âœ… å½’ä¸€åŒ–å®Œæˆ")
````

````python
# ============================================================
#            ğŸš€ è®­ç»ƒ LightGBM æ¨¡å‹
# ============================================================

print("ğŸš€ è®­ç»ƒ LightGBM æ¨¡å‹")
print("=" * 60)

# åˆ›å»º LightGBM æ•°æ®é›†
train_data = lgb.Dataset(X_train_norm, label=y_train)
valid_data = lgb.Dataset(X_valid_norm, label=y_valid, reference=train_data)

# æ¨¡å‹å‚æ•°
params = {
    "objective": "regression",     # å›å½’ä»»åŠ¡
    "metric": "mse",               # è¯„ä¼°æŒ‡æ ‡
    "num_leaves": 63,              # å¶å­èŠ‚ç‚¹æ•°
    "learning_rate": 0.05,         # å­¦ä¹ ç‡
    "feature_fraction": 0.8,       # ç‰¹å¾é‡‡æ ·
    "bagging_fraction": 0.8,       # æ ·æœ¬é‡‡æ ·
    "bagging_freq": 5,             # é‡‡æ ·é¢‘ç‡
    "verbose": -1,                 # é™é»˜æ¨¡å¼
}

print("ğŸ“‹ æ¨¡å‹å‚æ•°:")
for k, v in params.items():
    print(f"   {k}: {v}")

# è®­ç»ƒæ¨¡å‹
print("\nâ³ å¼€å§‹è®­ç»ƒ...")
model = lgb.train(
    params,
    train_data,
    num_boost_round=500,
    valid_sets=[valid_data],
    callbacks=[
        lgb.early_stopping(50),     # æ—©åœ
        lgb.log_evaluation(100),    # æ¯ 100 è½®æ‰“å°
    ],
)

print(f"\nâœ… è®­ç»ƒå®Œæˆ!")
print(f"   æœ€ä½³è¿­ä»£: {model.best_iteration}")
print(f"   æœ€ä½³éªŒè¯ MSE: {model.best_score['valid_0']['l2']:.8f}")
````

````python
# ============================================================
#            ğŸ“Š æ¨¡å‹è¯„ä¼° (Plotly äº¤äº’å¼)
# ============================================================
import plotly.graph_objs as go
from plotly.subplots import make_subplots

print("ğŸ“Š æ¨¡å‹è¯„ä¼°")
print("=" * 60)

# é¢„æµ‹
y_pred_train = model.predict(X_train_norm)
y_pred_valid = model.predict(X_valid_norm)

# IC (ä¿¡æ¯ç³»æ•°)
train_ic = np.corrcoef(y_pred_train, y_train)[0, 1]
valid_ic = np.corrcoef(y_pred_valid, y_valid)[0, 1]

print(f"ğŸ¯ IC (ä¿¡æ¯ç³»æ•°):")
print(f"   è®­ç»ƒé›† IC: {train_ic:.4f}")
print(f"   éªŒè¯é›† IC: {valid_ic:.4f}")

# IC è§£é‡Š
if valid_ic > 0.05:
    print(f"   âœ… IC > 0.05ï¼Œæ¨¡å‹æœ‰é¢„æµ‹èƒ½åŠ›")
elif valid_ic > 0.02:
    print(f"   âš ï¸ IC åœ¨ 0.02-0.05ï¼Œæ¨¡å‹é¢„æµ‹èƒ½åŠ›ä¸€èˆ¬")
else:
    print(f"   âŒ IC < 0.02ï¼Œæ¨¡å‹é¢„æµ‹èƒ½åŠ›è¾ƒå¼±")

# ç‰¹å¾é‡è¦æ€§
print("\nğŸ“ˆ Top 10 é‡è¦ç‰¹å¾:")
importance = pd.DataFrame({
    "feature": FEATURE_COLUMNS,
    "importance": model.feature_importance(importance_type="gain"),
}).sort_values("importance", ascending=False)

for i, row in importance.head(10).iterrows():
    bar = "â–ˆ" * int(row["importance"] / importance["importance"].max() * 20)
    print(f"   {row['feature']:12s} {bar} {row['importance']:.1f}")

# åˆ›å»º Plotly å¯è§†åŒ–
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        f'Prediction vs Actual (IC={valid_ic:.4f})',
        'Top 15 Feature Importance'
    )
)

# é¢„æµ‹ vs å®é™… (ä½¿ç”¨ Scattergl åŠ é€Ÿå¤§é‡ç‚¹çš„æ¸²æŸ“)
fig.add_trace(
    go.Scattergl(
        x=y_valid, y=y_pred_valid,
        mode='markers',
        marker=dict(size=3, opacity=0.3, color='royalblue'),
        name='Predictions'
    ),
    row=1, col=1
)

# å®Œç¾é¢„æµ‹çº¿
fig.add_trace(
    go.Scatter(
        x=[-0.1, 0.1], y=[-0.1, 0.1],
        mode='lines',
        line=dict(color='red', dash='dash'),
        name='Perfect'
    ),
    row=1, col=1
)

# ç‰¹å¾é‡è¦æ€§ (æ°´å¹³æŸ±çŠ¶å›¾)
top_n = 15
top_features = importance.head(top_n).iloc[::-1]  # ç¿»è½¬ä»¥ä¾¿æœ€é‡è¦çš„åœ¨ä¸Šé¢
fig.add_trace(
    go.Bar(
        x=top_features['importance'],
        y=top_features['feature'],
        orientation='h',
        marker_color='royalblue',
        name='Importance'
    ),
    row=1, col=2
)

fig.update_xaxes(title_text='Actual Return', row=1, col=1)
fig.update_yaxes(title_text='Predicted Return', row=1, col=1)
fig.update_xaxes(title_text='Importance (gain)', row=1, col=2)

fig.update_layout(
    height=450,
    showlegend=True,
    title_text='æ¨¡å‹è¯„ä¼°ç»“æœ',
    title_x=0.5
)

fig.show()

print("\nğŸ“Š è¯´æ˜:")
print("   - å·¦å›¾: é¢„æµ‹æ”¶ç›Šç‡ vs å®é™…æ”¶ç›Šç‡æ•£ç‚¹å›¾")
print("   - å³å›¾: Top 15 é‡è¦å› å­")
print("   - ç‚¹å‡»å›¾ä¾‹å¯ä»¥éšè—/æ˜¾ç¤ºå¯¹åº”ç³»åˆ—")
````

## 8.2 IC/ICIR å› å­æœ‰æ•ˆæ€§åˆ†æ (Qlib æ–¹æ³•)

> **æ¥æº**: [Qlib è¯„ä¼°æ–‡æ¡£](https://qlib.readthedocs.io/en/latest/component/report.html)

### ä»€ä¹ˆæ˜¯ IC å’Œ ICIRï¼Ÿ

| æŒ‡æ ‡ | å…¬å¼ | å«ä¹‰ | ä¼˜ç§€æ ‡å‡† |
|------|------|------|----------|
| **IC** | `corr(prediction, actual_return)` | é¢„æµ‹ä¸å®é™…æ”¶ç›Šçš„ç›¸å…³æ€§ | > 0.03 |
| **ICIR** | `mean(IC) / std(IC)` | IC çš„ç¨³å®šæ€§ | > 0.5 |

AlgVex çš„ `train_model.py` å·²å†…ç½® IC è®¡ç®— (ç¬¬ 199-201 è¡Œ)ã€‚

````python
# ============================================================
#            ğŸ“Š IC/ICIR å› å­åˆ†æ (è°ƒç”¨ç³»ç»Ÿç»„ä»¶)
# ============================================================
import sys, os
import numpy as np
import pandas as pd
from pathlib import Path
import lightgbm as lgb

from scripts.unified_features import (
    compute_unified_features, compute_label, FeatureNormalizer, FEATURE_COLUMNS
)

MODEL_DIR = Path.home() / ".algvex" / "models" / "qlib_alpha"
DATA_DIR = Path.home() / ".algvex" / "data" / "1h"

if MODEL_DIR.exists() and (DATA_DIR / "btcusdt.parquet").exists():
    model = lgb.Booster(model_file=str(MODEL_DIR / "lgb_model.txt"))
    normalizer = FeatureNormalizer()
    normalizer.load(str(MODEL_DIR / "normalizer.pkl"))
    
    df = pd.read_parquet(DATA_DIR / "btcusdt.parquet")
    test_df = df["2024-07-01":"2024-12-31"]
    
    features = compute_unified_features(test_df)
    labels = compute_label(test_df)
    
    valid_idx = features.dropna().index.intersection(labels.dropna().index)
    X = features.loc[valid_idx][FEATURE_COLUMNS]
    y = labels.loc[valid_idx]
    
    X_norm = normalizer.transform(X, strict=False)
    predictions = model.predict(X_norm.values)
    
    # è®¡ç®—æ•´ä½“ IC
    ic = np.corrcoef(predictions, y)[0, 1]
    
    # è®¡ç®—å‘¨åº¦ IC
    weekly_ics = []
    for week_start in pd.date_range(valid_idx.min(), valid_idx.max(), freq='7D'):
        week_end = week_start + pd.Timedelta(days=7)
        mask = (valid_idx >= week_start) & (valid_idx < week_end)
        if mask.sum() > 10:
            week_ic = np.corrcoef(predictions[mask.values], y.loc[valid_idx[mask]].values)[0, 1]
            weekly_ics.append(week_ic)
    
    icir = np.mean(weekly_ics) / (np.std(weekly_ics) + 1e-8) if weekly_ics else 0
    
    print("ğŸ“Š IC/ICIR åˆ†æç»“æœ")
    print("=" * 60)
    print(f"æ•´ä½“ IC: {ic:.4f}" + (" âœ…" if abs(ic) > 0.03 else " âš ï¸"))
    print(f"ICIR: {icir:.2f}" + (" âœ…" if icir > 0.5 else " âš ï¸"))
    print(f"\n(æ³¨: æ¨¡æ‹Ÿæ•°æ®çš„ IC é€šå¸¸æ¥è¿‘ 0ï¼Œè¿™æ˜¯æ­£å¸¸çš„)")
else:
    print("âš ï¸ è¯·å…ˆå®Œæˆæ¨¡å‹è®­ç»ƒ")
````

````python
# ============================================================
#            âœ… è‡ªåŠ¨åŒ–éªŒæ”¶æµ‹è¯•
# ============================================================
import sys
import os
from pathlib import Path
import pickle
import json

import numpy as np
import pandas as pd
import lightgbm as lgb

from scripts.unified_features import (
    compute_unified_features,
    FeatureNormalizer,
    FEATURE_COLUMNS,
)

print("âœ… AlgVex v10.0.4 éªŒæ”¶æµ‹è¯•")
print("=" * 60)

DATA_DIR = Path.home() / ".algvex" / "data" / "1h"
MODEL_DIR = Path.home() / ".algvex" / "models" / "qlib_alpha"

results = []

# ========== æµ‹è¯• 1: æ•°æ®å‡†å¤‡ ==========
print("\n[1/8] æ•°æ®å‡†å¤‡...")
try:
    btc_path = DATA_DIR / "btcusdt.parquet"
    eth_path = DATA_DIR / "ethusdt.parquet"
    
    if btc_path.exists() and eth_path.exists():
        btc_df = pd.read_parquet(btc_path)
        eth_df = pd.read_parquet(eth_path)
        
        # æ£€æŸ¥æ•°æ®é‡ (2å¹´çº¦ 17520 æ¡)
        if len(btc_df) >= 10000 and len(eth_df) >= 10000:
            print(f"   âœ… BTC: {len(btc_df):,} æ¡, ETH: {len(eth_df):,} æ¡")
            results.append(("æ•°æ®å‡†å¤‡", True))
        else:
            print(f"   âš ï¸ æ•°æ®é‡ä¸è¶³")
            results.append(("æ•°æ®å‡†å¤‡", False))
    else:
        print(f"   âŒ æ•°æ®æ–‡ä»¶ä¸å­˜åœ¨")
        results.append(("æ•°æ®å‡†å¤‡", False))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("æ•°æ®å‡†å¤‡", False))

# ========== æµ‹è¯• 2: ç‰¹å¾è®¡ç®— ==========
print("\n[2/8] ç‰¹å¾è®¡ç®—...")
try:
    features = compute_unified_features(btc_df.tail(100))
    
    if len(features.columns) == 59:
        print(f"   âœ… 59 å› å­è®¡ç®—æ­£ç¡®")
        results.append(("ç‰¹å¾è®¡ç®—", True))
    else:
        print(f"   âŒ å› å­æ•°é‡: {len(features.columns)}")
        results.append(("ç‰¹å¾è®¡ç®—", False))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("ç‰¹å¾è®¡ç®—", False))

# ========== æµ‹è¯• 3: æ¨¡å‹åŠ è½½ ==========
print("\n[3/8] æ¨¡å‹åŠ è½½...")
try:
    model_path = MODEL_DIR / "lgb_model.txt"
    normalizer_path = MODEL_DIR / "normalizer.pkl"
    
    if model_path.exists() and normalizer_path.exists():
        model = lgb.Booster(model_file=str(model_path))
        normalizer = FeatureNormalizer()
        normalizer.load(str(normalizer_path))
        
        print(f"   âœ… æ¨¡å‹å’Œå½’ä¸€åŒ–å™¨åŠ è½½æˆåŠŸ")
        results.append(("æ¨¡å‹åŠ è½½", True))
    else:
        print(f"   âŒ æ¨¡å‹æ–‡ä»¶ä¸å­˜åœ¨")
        results.append(("æ¨¡å‹åŠ è½½", False))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("æ¨¡å‹åŠ è½½", False))

# ========== æµ‹è¯• 4: å½’ä¸€åŒ– fit/transform ==========
print("\n[4/8] å½’ä¸€åŒ–å™¨...")
try:
    test_features = features.dropna().tail(50)
    
    # æµ‹è¯• transform
    norm_features = normalizer.transform(test_features, strict=True)
    
    if norm_features.shape == test_features.shape:
        print(f"   âœ… transform(strict=True) æ­£å¸¸")
        results.append(("å½’ä¸€åŒ–å™¨", True))
    else:
        print(f"   âŒ è¾“å‡ºå½¢çŠ¶ä¸åŒ¹é…")
        results.append(("å½’ä¸€åŒ–å™¨", False))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("å½’ä¸€åŒ–å™¨", False))

# ========== æµ‹è¯• 5: æ¨¡å‹é¢„æµ‹ ==========
print("\n[5/8] æ¨¡å‹é¢„æµ‹...")
try:
    predictions = model.predict(norm_features.values)
    
    if len(predictions) == len(norm_features):
        print(f"   âœ… é¢„æµ‹è¾“å‡º: {len(predictions)} æ¡")
        print(f"      å‡å€¼: {predictions.mean():.6f}")
        print(f"      èŒƒå›´: [{predictions.min():.6f}, {predictions.max():.6f}]")
        results.append(("æ¨¡å‹é¢„æµ‹", True))
    else:
        print(f"   âŒ è¾“å‡ºé•¿åº¦ä¸åŒ¹é…")
        results.append(("æ¨¡å‹é¢„æµ‹", False))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("æ¨¡å‹é¢„æµ‹", False))

# ========== æµ‹è¯• 6: ä¿¡å·ç”Ÿæˆ ==========
print("\n[6/8] ä¿¡å·ç”Ÿæˆ...")
try:
    threshold = 0.001
    long_signals = (predictions > threshold).sum()
    short_signals = (predictions < -threshold).sum()
    no_signals = len(predictions) - long_signals - short_signals
    
    print(f"   âœ… ä¿¡å·åˆ†å¸ƒ:")
    print(f"      å¤šå¤´: {long_signals} ({long_signals/len(predictions):.1%})")
    print(f"      ç©ºå¤´: {short_signals} ({short_signals/len(predictions):.1%})")
    print(f"      æ— ä¿¡å·: {no_signals} ({no_signals/len(predictions):.1%})")
    results.append(("ä¿¡å·ç”Ÿæˆ", True))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("ä¿¡å·ç”Ÿæˆ", False))

# ========== æµ‹è¯• 7: ä¸¥æ ¼æ¨¡å¼ ==========
print("\n[7/8] ä¸¥æ ¼æ¨¡å¼...")
try:
    # æµ‹è¯•æœ‰ NaN çš„æƒ…å†µ
    test_with_nan = features.head(10)  # å‰ 10 è¡Œå¯èƒ½æœ‰ NaN
    
    try:
        normalizer.transform(test_with_nan, strict=True)
        print(f"   âš ï¸ æœªæ£€æµ‹åˆ° NaN (å¯èƒ½æ•°æ®å·²ç»æœ‰æ•ˆ)")
        results.append(("ä¸¥æ ¼æ¨¡å¼", True))
    except ValueError as ve:
        if "NaN" in str(ve) or "ç¼ºå¤±" in str(ve):
            print(f"   âœ… strict=True æ­£ç¡®æ£€æµ‹ NaN")
            results.append(("ä¸¥æ ¼æ¨¡å¼", True))
        else:
            raise ve
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("ä¸¥æ ¼æ¨¡å¼", False))

# ========== æµ‹è¯• 8: å…ƒæ•°æ® ==========
print("\n[8/8] å…ƒæ•°æ®...")
try:
    metadata_path = MODEL_DIR / "metadata.json"
    
    if metadata_path.exists():
        with open(metadata_path) as f:
            metadata = json.load(f)
        
        print(f"   âœ… å…ƒæ•°æ®:")
        print(f"      ç‰ˆæœ¬: {metadata.get('version', 'N/A')}")
        print(f"      åˆ›å»ºæ—¶é—´: {metadata.get('created', 'N/A')[:19]}")
        print(f"      éªŒè¯é›† IC: {metadata.get('valid_ic', 'N/A')}")
        results.append(("å…ƒæ•°æ®", True))
    else:
        print(f"   âŒ å…ƒæ•°æ®æ–‡ä»¶ä¸å­˜åœ¨")
        results.append(("å…ƒæ•°æ®", False))
except Exception as e:
    print(f"   âŒ é”™è¯¯: {e}")
    results.append(("å…ƒæ•°æ®", False))

# ========== æ±‡æ€» ==========
print("\n" + "=" * 60)
print("ğŸ“‹ éªŒæ”¶ç»“æœæ±‡æ€»")
print("=" * 60)

passed = 0
for name, status in results:
    icon = "âœ…" if status else "âŒ"
    print(f"   {icon} {name}")
    if status:
        passed += 1

print(f"\n   é€šè¿‡: {passed}/{len(results)}")

if passed == len(results):
    print("\nğŸ‰ æ‰€æœ‰éªŒæ”¶æµ‹è¯•é€šè¿‡!")
    print("   å¯ä»¥è¿›å…¥ Paper Trading é˜¶æ®µ")
else:
    print(f"\nâš ï¸ {len(results) - passed} é¡¹æœªé€šè¿‡")
    print("   è¯·æ£€æŸ¥å¹¶ä¿®å¤é—®é¢˜åé‡è¯•")
````
