```python
#==============================================================================
#                              📁 数据配置
#==============================================================================

# Qlib 数据存储路径
# Windows 用户使用 r"路径" 格式，Mac/Linux 用户使用 "~/路径" 格式
DATA_PATH = r"F:\qlib_project\data\us_data"  # Windows 示例
# DATA_PATH = "~/.qlib/qlib_data/us_data"    # Mac/Linux 示例

#==============================================================================
#                              📅 时间配置
#==============================================================================

# 训练集时间范围：用于训练模型的历史数据
TRAIN_START = "2018-01-01"  # 训练开始日期
TRAIN_END = "2019-06-30"    # 训练结束日期（建议至少1年数据）

# 验证集时间范围：用于调参和早停的数据
VALID_START = "2019-07-01"  # 验证开始日期
VALID_END = "2019-12-31"    # 验证结束日期（建议3-6个月）

# 测试集时间范围：用于最终评估的数据（模拟未来）
TEST_START = "2020-01-01"   # 测试开始日期
TEST_END = "2020-06-30"     # 测试结束日期

#==============================================================================
#                              📊 股票配置
#==============================================================================

# 训练使用的股票数量（越多训练越慢，但模型可能更稳定）
# 建议：入门用50，正式研究用200-500
NUM_STOCKS = 50

# Top-K 策略每天选择的股票数量
# 数量越少集中度越高、波动越大；数量越多越分散、越稳定
TOP_K = 5

# 用于单股票演示的股票代码
STOCK_FOR_DEMO = "AAPL"  # 苹果公司

#==============================================================================
#                              🤖 模型选择
#==============================================================================

# 可选模型: "lightgbm", "xgboost", "catboost", "mlp", "gru", "lstm", "transformer"
# 注意: 部分模型需要额外安装依赖包
SELECTED_MODEL = "lightgbm"  # 默认使用 LightGBM（最稳定）

#==============================================================================
#                          🔧 LightGBM 模型参数
#==============================================================================

LIGHTGBM_CONFIG = {
    "num_boost_round": 500,        # 最大迭代轮数（树的数量）
    "early_stopping_rounds": 50,   # 早停轮数（验证集N轮不改善则停止）
    "learning_rate": 0.05,         # 学习率（越小越稳定，但训练更慢）
    "num_leaves": 64,              # 叶子节点数（越大模型越复杂）
    "max_depth": 8,                # 树的最大深度（限制复杂度）
    "lambda_l1": 200,              # L1正则化系数（防止过拟合）
    "lambda_l2": 200,              # L2正则化系数（防止过拟合）
    "colsample_bytree": 0.8,       # 每棵树使用的特征比例
    "subsample": 0.8,              # 每棵树使用的样本比例
}

#==============================================================================
#                          🔧 XGBoost 模型参数
#==============================================================================

XGBOOST_CONFIG = {
    "n_estimators": 500,           # 树的数量
    "early_stopping_rounds": 50,   # 早停轮数
    "learning_rate": 0.05,         # 学习率
    "max_depth": 8,                # 最大深度
    "reg_alpha": 200,              # L1正则化
    "reg_lambda": 200,             # L2正则化
    "colsample_bytree": 0.8,       # 特征采样比例
    "subsample": 0.8,              # 样本采样比例
}

#==============================================================================
#                          🔧 CatBoost 模型参数
#==============================================================================

CATBOOST_CONFIG = {
    "iterations": 500,             # 迭代次数
    "early_stopping_rounds": 50,   # 早停轮数
    "learning_rate": 0.05,         # 学习率
    "depth": 8,                    # 树深度
    "l2_leaf_reg": 200,            # L2正则化
    "random_seed": 42,             # 随机种子
}

#==============================================================================
#                          🔧 神经网络模型参数 (MLP/GRU/LSTM)
#==============================================================================

NN_CONFIG = {
    "d_model": 64,                 # 隐藏层维度
    "batch_size": 2048,            # 批量大小
    "dropout": 0.5,                # Dropout比例（防止过拟合）
    "n_epochs": 100,               # 训练轮数
    "lr": 0.001,                   # 学习率
    "early_stop": 10,              # 早停轮数
    "GPU": 0,                      # GPU编号（-1表示使用CPU）
}

#==============================================================================
#                          🔧 完整回测配置
#==============================================================================

BACKTEST_CONFIG = {
    "account": 1000000,            # 初始资金（美元）
    "topk": 30,                    # 持仓股票数量
    "n_drop": 5,                   # 每次调仓卖出数量
    "open_cost": 0.0005,           # 买入手续费率（0.05%）
    "close_cost": 0.0015,          # 卖出手续费率（0.15%，含印花税）
    "min_cost": 5,                 # 最低手续费（美元）
}

#==============================================================================
#                              🎯 评估配置
#==============================================================================

RISK_FREE_RATE = 0.02    # 无风险利率（年化），用于计算夏普比率
TRADING_DAYS = 252       # 每年交易日数（美股252天）

#==============================================================================
print("✅ 配置加载完成！")
print(f"📁 数据路径: {DATA_PATH}")
print(f"📊 股票数量: {NUM_STOCKS}")
print(f"🤖 选择模型: {SELECTED_MODEL}")
print(f"📅 测试期间: {TEST_START} ~ {TEST_END}")
print(f"💰 初始资金: ${BACKTEST_CONFIG['account']:,}")
```

```python
import warnings
warnings.filterwarnings('ignore')

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
pd.set_option('display.max_columns', 20)

print("✅ 基础库导入成功！")
```

```python
import qlib
from qlib.constant import REG_US

def safe_init_qlib(data_path, region=REG_US):
    """安全初始化 Qlib，防止重复初始化报错"""
    try:
        from qlib.data import D
        D.calendar(freq="day")[:1]
        print("⚠️ Qlib 已经初始化过了，跳过重复初始化")
        return True
    except:
        pass
    
    expanded_path = os.path.expanduser(data_path)
    if not os.path.exists(expanded_path):
        print(f"❌ 数据路径不存在: {expanded_path}")
        print("请下载数据: python -m qlib.run.get_data qlib_data --target_dir ~/.qlib/qlib_data/us_data --region us")
        return False
    
    try:
        qlib.init(provider_uri=data_path, region=region)
        print(f"✅ Qlib 初始化成功！数据路径: {data_path}")
        return True
    except Exception as e:
        print(f"❌ Qlib 初始化失败: {e}")
        return False

qlib_ready = safe_init_qlib(DATA_PATH)
```

# Part 4: 多模型训练与对比

## 4.1 准备数据集

```python
from qlib.data.dataset import DatasetH
from qlib.contrib.data.handler import Alpha158

train_stocks = all_instruments[:NUM_STOCKS]

print(f"📊 准备数据集: {len(train_stocks)} 只股票")
print(f"训练期: {TRAIN_START} ~ {TRAIN_END}")
print(f"验证期: {VALID_START} ~ {VALID_END}")
print(f"测试期: {TEST_START} ~ {TEST_END}")

handler = Alpha158(
    instruments=train_stocks,
    start_time=TRAIN_START,
    end_time=TEST_END,
    fit_start_time=TRAIN_START,
    fit_end_time=TRAIN_END,
)

dataset = DatasetH(
    handler=handler,
    segments={
        "train": (TRAIN_START, TRAIN_END),
        "valid": (VALID_START, VALID_END),
        "test": (TEST_START, TEST_END),
    }
)

train_data = dataset.prepare("train", col_set=["feature", "label"])
test_data = dataset.prepare("test", col_set=["feature", "label"])
print(f"✅ 训练集: {train_data.shape}, 测试集: {test_data.shape}")
```

## 4.2 模型工厂函数

```python
def create_model(model_name):
    """
    根据模型名称创建对应的模型实例
    
    Args:
        model_name: "lightgbm", "xgboost", "catboost", "mlp", "gru", "lstm"
    
    Returns:
        model: Qlib 模型实例
    """
    if model_name == "lightgbm":
        from qlib.contrib.model.gbdt import LGBModel
        return LGBModel(
            loss="mse",
            num_boost_round=LIGHTGBM_CONFIG["num_boost_round"],
            early_stopping_rounds=LIGHTGBM_CONFIG["early_stopping_rounds"],
            num_leaves=LIGHTGBM_CONFIG["num_leaves"],
            learning_rate=LIGHTGBM_CONFIG["learning_rate"],
            max_depth=LIGHTGBM_CONFIG["max_depth"],
            lambda_l1=LIGHTGBM_CONFIG["lambda_l1"],
            lambda_l2=LIGHTGBM_CONFIG["lambda_l2"],
            colsample_bytree=LIGHTGBM_CONFIG["colsample_bytree"],
            subsample=LIGHTGBM_CONFIG["subsample"],
        )
    
    elif model_name == "xgboost":
        try:
            from qlib.contrib.model.xgboost import XGBModel
            return XGBModel(**XGBOOST_CONFIG)
        except ImportError:
            print("❌ 请安装 XGBoost: pip install xgboost")
            return None
    
    elif model_name == "catboost":
        try:
            from qlib.contrib.model.catboost_model import CatBoostModel
            return CatBoostModel(**CATBOOST_CONFIG)
        except ImportError:
            print("❌ 请安装 CatBoost: pip install catboost")
            return None
    
    elif model_name == "mlp":
        try:
            from qlib.contrib.model.pytorch_nn import DNNModel
            return DNNModel(**NN_CONFIG)
        except ImportError:
            print("❌ 请安装 PyTorch: pip install torch")
            return None
    
    elif model_name == "gru":
        try:
            from qlib.contrib.model.pytorch_gru import GRU
            return GRU(**NN_CONFIG)
        except ImportError:
            print("❌ 请安装 PyTorch: pip install torch")
            return None
    
    elif model_name == "lstm":
        try:
            from qlib.contrib.model.pytorch_lstm import LSTM
            return LSTM(**NN_CONFIG)
        except ImportError:
            print("❌ 请安装 PyTorch: pip install torch")
            return None
    
    else:
        print(f"❌ 未知模型: {model_name}")
        return None

print("✅ 模型工厂函数已定义")
print("支持的模型: lightgbm, xgboost, catboost, mlp, gru, lstm")
```

## 4.3 训练选定的模型

```python
print(f"🚀 开始训练模型: {SELECTED_MODEL}")

model = create_model(SELECTED_MODEL)

if model is not None:
    model.fit(dataset)
    print("✅ 模型训练完成！")
    
    # 生成预测
    predictions = model.predict(dataset, segment="test")
    print(f"预测结果: {len(predictions)} 条")
else:
    print("❌ 模型创建失败")
```

# Part 5: 简单回测与评估

## 5.1 对齐预测与标签

```python
test_labels = test_data["label"].squeeze()
common_idx = predictions.index.intersection(test_labels.index)
pred_aligned = predictions.loc[common_idx]
label_aligned = test_labels.loc[common_idx]

print(f"对齐后数据量: {len(common_idx)}")
```

## 5.2 IC 分析

```python
comparison = pd.DataFrame({'prediction': pred_aligned, 'actual': label_aligned})

daily_corr = comparison.groupby(level='datetime').apply(
    lambda x: x['prediction'].corr(x['actual'])
).dropna()

ic_mean = daily_corr.mean()
ic_std = daily_corr.std()
ir = ic_mean / ic_std if ic_std > 0 else 0

print(f"📊 IC 分析:")
print(f"  IC 均值: {ic_mean:.4f}")
print(f"  IC 标准差: {ic_std:.4f}")
print(f"  IR: {ir:.4f}")
print(f"  IC > 0 比例: {(daily_corr > 0).mean():.2%}")
```

## 5.3 分组回测

```python
def assign_group(x):
    try:
        return pd.qcut(x['prediction'], q=5, 
                       labels=['G1_Low', 'G2', 'G3', 'G4', 'G5_High'], 
                       duplicates='drop')
    except:
        return pd.Series([np.nan] * len(x), index=x.index)

comparison['group'] = comparison.groupby(level='datetime').apply(assign_group).droplevel(0)
comparison_valid = comparison.dropna(subset=['group'])

group_returns = comparison_valid.groupby(['datetime', 'group'])['actual'].mean().unstack()
print("各组平均日收益:")
print(group_returns.mean())
```

```python
# 可视化分组收益
cumulative_returns = (1 + group_returns).cumprod()

plt.figure(figsize=(14, 6))
color_map = {'G1_Low': 'red', 'G2': 'orange', 'G3': 'gray', 'G4': 'lightgreen', 'G5_High': 'green'}

for col in cumulative_returns.columns:
    plt.plot(cumulative_returns.index, cumulative_returns[col], 
             label=col, linewidth=2, color=color_map.get(col, 'blue'))

plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.title('Cumulative Returns by Prediction Quintile')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5.4 Top-K 策略

```python
K = TOP_K

def get_topk_return(group):
    return group.nlargest(K, 'prediction')['actual'].mean()

strategy_returns = comparison_valid.groupby(level='datetime').apply(get_topk_return)
benchmark_returns = comparison_valid.groupby(level='datetime')['actual'].mean()

print(f"Top-{K} 策略日均收益: {strategy_returns.mean():.4%}")
print(f"基准日均收益: {benchmark_returns.mean():.4%}")
print(f"超额收益: {(strategy_returns.mean() - benchmark_returns.mean()):.4%}")
```

## 5.5 综合评估报告

```python
cum_strategy = (1 + strategy_returns).cumprod()
peak = cum_strategy.expanding().max()
drawdown = (cum_strategy - peak) / peak

total_return = cum_strategy.iloc[-1] - 1
annual_return = strategy_returns.mean() * TRADING_DAYS
annual_vol = strategy_returns.std() * np.sqrt(TRADING_DAYS)
sharpe = (annual_return - RISK_FREE_RATE) / annual_vol if annual_vol > 0 else 0
max_dd = drawdown.min()
win_rate = (strategy_returns > 0).mean()

print("=" * 60)
print("📊 策略综合评估报告")
print("=" * 60)
print(f"\n【收益指标】")
print(f"  总收益率: {total_return:.2%}")
print(f"  年化收益率: {annual_return:.2%}")
print(f"\n【风险指标】")
print(f"  年化波动率: {annual_vol:.2%}")
print(f"  最大回撤: {max_dd:.2%}")
print(f"  夏普比率: {sharpe:.2f}")
print(f"\n【交易指标】")
print(f"  胜率: {win_rate:.2%}")
print(f"\n【预测能力】")
print(f"  IC 均值: {ic_mean:.4f}")
print(f"  IR: {ir:.4f}")
print("=" * 60)

if sharpe >= 2.0 and ic_mean >= 0.03:
    grade = "A (优秀)"
elif sharpe >= 1.5:
    grade = "B+ (良好)"
elif sharpe >= 1.0:
    grade = "B (中等)"
else:
    grade = "C (需改进)"
print(f"📈 综合评级: {grade}")
```

# Part 6: Qlib 完整回测框架

## 6.1 使用 Qlib 内置回测引擎

Qlib 提供完整的回测框架，包括：
- **TopkDropoutStrategy**: Top-K 持仓策略
- **Backtest**: 模拟交易执行
- **RiskAnalysis**: 风险指标分析

```python
from qlib.contrib.strategy import TopkDropoutStrategy
from qlib.contrib.evaluate import backtest_daily, risk_analysis

print("📊 配置 Qlib 完整回测...")
print(f"  持仓数量: {BACKTEST_CONFIG['topk']}")
print(f"  每次调仓卖出: {BACKTEST_CONFIG['n_drop']}")
print(f"  初始资金: ${BACKTEST_CONFIG['account']:,}")
print(f"  买入费率: {BACKTEST_CONFIG['open_cost']:.2%}")
print(f"  卖出费率: {BACKTEST_CONFIG['close_cost']:.2%}")
```

```python
# 创建策略
strategy = TopkDropoutStrategy(
    signal=predictions,                    # 模型预测信号
    topk=BACKTEST_CONFIG["topk"],         # 持仓数量
    n_drop=BACKTEST_CONFIG["n_drop"],     # 每次卖出数量
)

print("✅ 策略创建成功！")
```

```python
# 执行回测
print("🚀 开始执行回测...")

try:
    portfolio_metric_dict, indicator_dict = backtest_daily(
        start_time=TEST_START,
        end_time=TEST_END,
        strategy=strategy,
        account=BACKTEST_CONFIG["account"],
        exchange_kwargs={
            "freq": "day",
            "limit_threshold": None,  # 美股无涨跌停
            "deal_price": "close",
            "open_cost": BACKTEST_CONFIG["open_cost"],
            "close_cost": BACKTEST_CONFIG["close_cost"],
            "min_cost": BACKTEST_CONFIG["min_cost"],
        },
    )
    print("✅ 回测完成！")
    backtest_success = True
except Exception as e:
    print(f"❌ 回测失败: {e}")
    backtest_success = False
```

```python
# 分析回测结果
if backtest_success:
    report = risk_analysis(portfolio_metric_dict.get("return", portfolio_metric_dict))
    
    print("\n" + "=" * 60)
    print("📊 Qlib 回测结果分析")
    print("=" * 60)
    print(report)
    print("=" * 60)
```

# Part 7: 自定义因子开发

## 7.1 Qlib 表达式语法

Qlib 支持使用表达式语法定义因子：

| 函数 | 说明 | 示例 |
|------|------|------|
| `$close` | 收盘价 | `$close` |

| `Mean($close, N)` | N日均值 | `Mean($close, 20)` |
| `Std($close, N)` | N日标准差 | `Std($close, 20)` |
| `Max($high, N)` | N日最高 | `Max($high, 20)` |
| `Corr($x, $y, N)` | N日相关性 | `Corr($close, $volume, 5)` |

## 7.2 自定义因子处理器

```python
from qlib.contrib.data.handler import DataHandlerLP

class MyAlphaHandler(DataHandlerLP):
    """
    自定义因子处理器示例
    继承 DataHandlerLP 并定义自己的因子
    """
    
    def __init__(self, instruments, start_time, end_time, **kwargs):
        # 定义特征因子
        feature_config = [
            ("KMID", "($close - $open) / $open"),
            ("KLEN", "($high - $low) / $open"),
            ("MOM5", "$close / Ref($close, 5) - 1"),
            ("MOM10", "$close / Ref($close, 10) - 1"),
            ("MOM20", "$close / Ref($close, 20) - 1"),
            ("VOL5", "Std($close / Ref($close, 1) - 1, 5)"),
            ("VOL20", "Std($close / Ref($close, 1) - 1, 20)"),
            ("MA_BIAS", "($close - Mean($close, 20)) / Mean($close, 20)"),
            ("VWAP_RATIO", "Sum($close * $volume, 5) / Sum($volume, 5) / $close"),
            ("VP_CORR", "Corr($close, $volume, 5)"),
        ]
        
        # 定义标签（未来收益率）
        label_config = [
            ("LABEL0", "Ref($close, -2) / Ref($close, -1) - 1"),
        ]
        
        data_loader = {
            "class": "QlibDataLoader",
            "kwargs": {
                "config": {
                    "feature": feature_config,
                    "label": label_config,
                },
            },
        }
        
        super().__init__(
            instruments=instruments,
            start_time=start_time,
            end_time=end_time,
            data_loader=data_loader,
            **kwargs
        )

print("✅ 自定义因子处理器已定义")
print("使用方法: handler = MyAlphaHandler(instruments=..., start_time=..., end_time=...)")
```

## 8.1 过拟合检测

```python
def check_overfitting(train_ic, test_ic):
    """
    检测模型是否过拟合
    
    Args:
        train_ic: 训练集 IC
        test_ic: 测试集 IC
    """
    ratio = test_ic / train_ic if train_ic > 0 else 0
    
    print(f"\n📊 过拟合检测:")
    print(f"  训练集 IC: {train_ic:.4f}")
    print(f"  测试集 IC: {test_ic:.4f}")
    print(f"  比率 (测试/训练): {ratio:.2%}")
    
    if ratio >= 0.8:
        print("  ✅ 正常：模型泛化良好")
    elif ratio >= 0.5:
        print("  ⚠️ 轻微过拟合：建议增加正则化")
    else:
        print("  ❌ 严重过拟合：需要重新调参")

# 使用示例
# check_overfitting(train_ic=0.08, test_ic=0.04)
```

## 8.2 回测陷阱检查清单

| 陷阱 | 描述 | 检测方法 |
|------|------|----------|
| **未来信息泄露** | 使用了未来数据 | 检查因子是否只用历史数据 |
| **幸存者偏差** | 只用现存股票 | 使用包含退市股票的数据 |
| **过度拟合** | 训练集表现远好于测试集 | 比较训练/测试 IC |
| **忽略交易成本** | 假设零成本 | 添加 open_cost/close_cost |
| **流动性忽略** | 假设无限流动性 | 限制持仓占比 |
